```python
import numpy as np
import matplotlib.pyplot as  plt
import pandas as pd
```

```python
df=pd.read_csv('crop_yield_m.csv')
df
```

```python
df.info()
```

```python
X = df.drop(columns=['Yield','Fertilizer','Pesticide'])
Y = df[['Yield','Fertilizer','Pesticide']] 
print(X.shape,Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape,Y_train.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
X_train['Crop'] = label_encoder.fit_transform(X_train['Crop'])
X_train['Season'] = label_encoder.fit_transform(X_train['Season'])
X_train['State'] = label_encoder.fit_transform(X_train['State'])

X_test['Crop'] = label_encoder.fit_transform(X_test['Crop'])
X_test['Season'] = label_encoder.fit_transform(X_test['Season'])
X_test['State'] = label_encoder.fit_transform(X_test['State'])

```

```python
from sklearn.ensemble import RandomForestRegressor
regressor =RandomForestRegressor(n_estimators=100,random_state=42,criterion='squared_error')
regressor.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_absolute_error
Y_pred=regressor.predict(X_train)
print(f'variance = {mean_absolute_error(Y_train,Y_pred)}')
Y_pred=regressor.predict(X_test)
print(f'bias = {mean_absolute_error(Y_test,Y_pred)}')
```

```python
import pickle
with open('Crop_Yeild_Predict.pkl', 'wb') as f:
   pickle.dump(regressor,f)
```
